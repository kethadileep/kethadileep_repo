```python
# MRO
print ('=================MRO-inheritance=====================')
class A:
    def sample(self):
        print('class A')


class B:
    def sample(self):
        print('class B')


class C(A, B):
    def sample(self):
        print('class C')


class D(C,B):
    pass


# obj = D()
print(D.mro())
```

```python
# Single Inheritance
# When a child class inherits only a single parent class.
class Parent:
    def func1(self):
        print("This is function from Parent")


class Child(Parent):
    def func2(self):
        print("This is function from Child")


ob = Child()
ob.func1()
ob.func2()
```

```python
# # Multiple Inheritance
# # When a child class inherits from more than one parent class.
class Parent:
   def func1(self):
        print("this is function from Parent1")
class Parent2:
   def func2(self):
        print("this is function from Parent2")
class Child(Parent , Parent2):
    def func3(self):
        print("this is function Child")

ob = Child()
ob.func1()
ob.func2()
ob.func3()
```

```python
# # Multilevel Inheritance
# # When a child class becomes a parent class for another child class.
class Parent:
      def func1(self):
          print("this is function from Parent")
class Child(Parent):
      def func2(self):
          print("this is function from Child1")
class Child2(Child):
      def func3(self):
          print ("this is function from Child2")
ob = Child2()
ob.func1()
ob.func2()
ob.func3()
```

```python
# # Hierarchical Inheritance
# # Hierarchical inheritance involves multiple inheritance from the same base or parent class.
class Parent:
      def func1(self):
          print("this is function from Parent")
class Child(Parent):
      def func2(self):
          print("this is function from child1")
class Child2(Parent):
      def func3(self):
          print("this is function from child2")

ob = Child()
ob1 = Child2()
ob.func1()
ob.func2()

```

```python
# # Hybrid Inheritance
# # Hybrid inheritance involves multiple inheritance taking place in a single program.

class Parent:
     def func1(self):
         print("this is function Parent")

class Child(Parent):
     def func2(self):
         print("this is function Child")

class Child1(Parent):
     def func3(self):
         print(" this is function Child1")

class Child3(Parent , Child1):
     def func4(self):
         print(" this is function from child3")

ob = Child3()
ob.func1()
```

```python
class Taxi:

    def __init__(self, model, capacity, variant):
        self.__model = model      # __model is private to Taxi class
        self.__capacity = capacity
        self.__variant = variant

    def getModel(self):          # getmodel() is accessible outside the class
        return self.__model

    def getCapacity(self):         # getCapacity() function is accessible to class Vehicle
        return self.__capacity

    def setCapacity(self, capacity):  # setCapacity() is accessible outside the class
        self.__capacity = capacity

    def getVariant(self):         # getVariant() function is accessible to class Vehicle
        return self.__variant

    def setVariant(self, variant):  # setVariant() is accessible outside the class
        self.__variant = variant

class Vehicle(Taxi):

    def __init__(self, model, capacity, variant, color):
        # call parent constructor to set model and color
        super().__init__(model, capacity, variant)
        self.__color = color

    def vehicleInfo(self):
        return self.getModel() + " " + self.getVariant() + " in " + self.__color + " with " + self.getCapacity() + " seats"

# In method getInfo we can call getmodel(), getCapacity() as they are
# accessible in the child class through inheritance

v1 = Vehicle("i20 Active", "4", "SX", "Bronze")
print(v1.vehicleInfo())
print(v1.getModel()) # Vehicle has no method getModel() but it is accessible via Vehicle class

v2 = Vehicle("Fortuner", "7", "MT2755", "White")
print(v2.vehicleInfo())
print(v2.getModel()) # Vehicle has no method getModel() but it is accessible via Vehicle class


# output
# i20 Active SX in Bronze with 4 seats
# i20 Active
# Fortuner MT2755 in White with 7 seats
# Fortuner
```

```python
class Emp():
    def __init__(self, id, name, Add):
        self.id = id
        self.name = name
        self.Add = Add

# Class freelancer inherits EMP
class Freelance(Emp):
    def __init__(self, id, name, Add, Emails):
        super().__init__(id, name, Add)
        self.Emails = Emails

Emp_1 = Freelance(103, "Suraj kr gupta", "Noida" , "KKK@gmails")
print('The ID is:', Emp_1.id)
print('The Name is:', Emp_1.name)
print('The Address is:', Emp_1.Add)
print('The Emails is:', Emp_1.Emails)
```

```python
# mro, super
class A:
    def age(self):
        print("Age is 21")
class B:
    def age(self):
        print("Age is 23")
class C(A, B):
    def age(self):
        super(C, self).age()

c = C()
print(C.__mro__)
print(C.mro())
```

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * self.length + 2 * self.width

# Here we declare that the Square class inherits from the Rectangle class
class Square(Rectangle):
    def __init__(self, length):
        super().__init__(length, length)
```

```python
class Square(Rectangle):
    def __init__(self, length):
        super().__init__(length, length)

class Cube(Square):
    def surface_area(self):
        face_area = super().area()
        return face_area * 6

    def volume(self):
        face_area = super().area()
        return face_area * self.length
```
